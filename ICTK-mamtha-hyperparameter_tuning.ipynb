```python
#import modules
import pandas as pd
import numpy as np
```

```python
#reading csv
df=pd.read_csv('Downloads/credit_card_default.csv')
df
```

##    Preprocessing

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.info()
```

```python
#convert uppercase into lowercase
df= df.rename(columns = lambda x: x.lower())
```

```python
df['sex'].value_counts()
```

```python
df['education'].value_counts()
```

```python
df['marriage'].value_counts()
```

```python
df['male'] = (df['sex']==1).astype(int)
df
```

```python
df['grad_school'] = (df['education']==1).astype(int)
df
```

```python
df['university'] = (df['education']==2).astype(int)
df
```

```python
df['married'] = (df['marriage']==1).astype(int)
df
```

```python
y=df['default payment next month']
X=df.drop(['default payment next month'],axis=1)
X.describe()
```

```python
#split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
#naive bayes
from sklearn.naive_bayes import GaussianNB
nb_clf = GaussianNB()
nb_clf.fit(X_train, y_train)
nb_clf.score(X_test, y_test)
```

```python
Y_pred = nb_clf.predict(X_test)

from sklearn.metrics import confusion_matrix

cm=confusion_matrix(y_test, Y_pred)
cm
```

```python
from sklearn.metrics import precision_score, accuracy_score, recall_score, f1_score
accuracy = accuracy_score(y_test, Y_pred)
accuracy*100
```

##  Knn

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
knn=KNeighborsClassifier()
# Set up the parameter grids
knn_param_grid = {'n_neighbors': [3, 5, 7, 9]}
```

```python
# Set up the GridSearchCV objects
knn_grid = GridSearchCV(knn, knn_param_grid, cv=5)
```

```python
knn_grid.fit(X_train,y_train)
```

```python
Y_pred=knn_grid.predict(X_test)
```

```python
# Print the best parameters and scores for each model
print('kNN:')
print(knn_grid.best_params_)
print(knn_grid.best_score_)
```

```python
from sklearn import metrics
("knn accuracy=",metrics.accuracy_score(y_test,Y_pred)*100)
```

Perform Classification Using SVM

```python
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, Y_pred))
```
